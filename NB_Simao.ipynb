```python
positions = df['position'].value_counts()
plt.figure(figsize=(6,5))
colors=['green','red', 'black','pink']
plt.bar(positions.index, positions.values, width=0.4, color=colors)
plt.title("Assists per Position",fontweight = 'bold')
plt.xlabel("Position")
plt.ylabel("Number")
plt.show()
```

```python
df2=df[:5]
plt.figure(figsize=(6,5))
colors=['green','red', 'black','gold','pink']
plt.bar(df2['player_name'], df2['assists'], width=0.4, color=colors)
plt.title("Assists per player",fontweight = 'bold')
plt.xlabel("Player")
plt.ylabel("Numberof assists")
plt.show()
```
