```python
import pandas as pd
```

```python
cust_info=pd.read_csv('customer_info.csv')
cust_info=cust_info.drop(columns=['customer_id'])
cust_info.head(10)
```

```python
cust_info.info()
```

```python
# 1. Your clean list of degrees
degrees = ['Bsc.', 'Msc.', 'Phd.']

# 2. Make sure the names are stripped of hidden spaces
cust_info['customer_name'] = cust_info['customer_name'].str.strip()

# 3. Create the regex pattern
pattern = r'^(' + '|'.join(degrees) + ')'

# 4. CRITICAL STEP: Use .str.extract() so pandas pulls out the matching text
cust_info['education'] = cust_info['customer_name'].str.extract(pattern)
cust_info.head(10)
```
